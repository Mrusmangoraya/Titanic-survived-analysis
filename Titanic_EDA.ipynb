### Visualized and analiyzed

```python
sns.countplot(data=df,x="survived")
```

```python
sns.barplot(data=df,x="survived",y="fare",hue="who")
```

```python
# data distribution
sns.histplot(data=df,x="age",bins=30)
```

```python
df["age"].hist(bins=30, color="darkred",alpha=0.75)
```

```python
sns.boxplot(data=df,x="survived",y="age")
```

```python
plt.figure(figsize=(6,3))
sns.boxplot(data=df,x="survived",y="age",hue="who")
```

```python
plt.figure(figsize=(8,4))
sns.scatterplot(x="age",y="fare",data=df, hue="pclass")
plt.ylim(0,300)
```

```python
sns.pairplot(df, hue="survived")
```

```python
# stacked barchart/plot
df.groupby(["pclass","who"])["fare"].mean().unstack().plot(kind="bar",stacked="True")
```

```python
sns.swarmplot(data=df, x="survived",y="age", hue="who")
```

```python
sns.catplot(x="pclass",y="fare",hue="embarked",kind="point",data=df)
```

```python
# how to add column and make age group
sns.histplot(data=df, x="age")
```

```python
# grouping
df.columns
```

Make the age group in the data

```python
df["age_groups"]=pd.cut(df["age"],[0,10,18,30,50,86],labels=["kids","child","youth","middle aged","old"])
df.head()
```

```python
df.columns
```

```python
df.groupby("age_groups")["survived"].mean().plot(kind="bar")
```

Pair plot between different variables

```python
sns.pairplot(df,hue="survived",vars=["age","fare","pclass"])
```

## Ratio between death and survived 

```python
survived_counts = df['survived'].value_counts()

plt.figure(figsize=(4, 4))
plt.pie(survived_counts, labels=survived_counts.index, autopct='%1.1f%%')
plt.title("Ratio of death and survived ")
```
